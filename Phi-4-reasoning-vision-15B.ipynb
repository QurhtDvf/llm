# Phi-4-reasoning-vision-15B -- テキスト推論 (llama.cpp CPU)

Microsoft released the multimodal model Phi-4-reasoning-vision-15B on March 4, 2026, with an official recommended specification of a 40GB+ GPU.
I adapted it to run in a CPU-only environment by dropping the image modality and focusing on text reasoning only, enabling it to run with llama-server and an OpenAI-compatible API.
 画像入力なし・テキストのみで動作。
think/nothink モードの切り替え、数学・論理推論などが使えます。

| 項目 | 内容 |
|------|------|
| モデル | Q4_K_M GGUF (CPU) |
| 画像入力 | ❌ 非対応（phi4 projector未対応のため） |
| テキスト推論 | ✅ think/nothink モード対応 |


## 1. Setup


```python
!pip install -q huggingface_hub psutil requests
import psutil, time, subprocess, requests, textwrap
from huggingface_hub import hf_hub_download
print('OK')
```

```python
REPO       = 'DevQuasar/microsoft.Phi-4-reasoning-vision-15B-GGUF'
MODEL_FILE = 'microsoft.Phi-4-reasoning-vision-15B.f16.gguf.Q4_K_M.gguf'
if psutil.virtual_memory().total / 1e9 < 12:
    MODEL_FILE = 'microsoft.Phi-4-reasoning-vision-15B.f16.gguf.Q2_K.gguf'
model_path = hf_hub_download(repo_id=REPO, filename=MODEL_FILE, local_dir='/content/models')
print(model_path)
```

## 2. llama.cpp ビルド（初回のみ）


```python
%%bash
if [ -f /content/llama.cpp/build/bin/llama-server ]; then
    echo 'Already built.'; exit 0
fi
git clone --depth=1 https://github.com/ggml-org/llama.cpp /content/llama.cpp
cd /content/llama.cpp
cmake -B build -DGGML_CUDA=OFF -DGGML_METAL=OFF \
    -DLLAMA_BUILD_TESTS=OFF -DLLAMA_BUILD_EXAMPLES=ON -DCMAKE_BUILD_TYPE=Release
cmake --build build --target llama-server -j$(nproc)
echo 'Done.' && ls -lh build/bin/llama-server
```

## 3. サーバー起動


```python
subprocess.run(['pkill', '-f', 'llama-server'], capture_output=True)
time.sleep(2)

SERVER_PORT = 8088
LOG_FILE = '/content/llama_server.log'

with open(LOG_FILE, 'w') as log:
    server_proc = subprocess.Popen(
        [
            '/content/llama.cpp/build/bin/llama-server',
            '-m',     model_path,
            '--host', '127.0.0.1',
            '--port', str(SERVER_PORT),
            '-c',     '4096',
            '-t',     '4',
            '-ngl',   '0',
        ],
        stdout=log, stderr=log,
    )

print(f'PID: {server_proc.pid}')
for i in range(120):
    time.sleep(2)
    if server_proc.poll() is not None:
        with open(LOG_FILE) as f: print(f.read()[-1000:])
        raise RuntimeError(f'Server exited (rc={server_proc.returncode})')
    try:
        r = requests.get(f'http://localhost:{SERVER_PORT}/health', timeout=2)
        if r.status_code == 200:
            print(f'Ready! ({i*2}s)  RAM={psutil.virtual_memory().used/1e9:.1f}GB')
            break
    except:
        pass
    if i % 10 == 9:
        print(f'  {i*2}s... RAM={psutil.virtual_memory().used/1e9:.1f}GB')
else:
    raise RuntimeError('Timeout')
```

## 4. 推論関数


```python
NOTHINK = '<|dummy_84|>'
SERVER_URL = f'http://localhost:{SERVER_PORT}'

def ask(user_text, think=True, max_tokens=400, system=None):
    system = system or 'You are Phi, a helpful assistant trained by Microsoft.'
    think_token = '' if think else NOTHINK
    payload = {
        'model': 'phi4',
        'messages': [
            {'role': 'system', 'content': system},
            {'role': 'user',   'content': f'{think_token}\n{user_text}'},
        ],
        'max_tokens': max_tokens,
        'temperature': 0.0,
        'stop': ['<|end|>', '<|user|>'],
    }
    r = requests.post(f'{SERVER_URL}/v1/chat/completions', json=payload, timeout=600)
    r.raise_for_status()
    return r.json()['choices'][0]['message']['content']


def show(response, show_thinking=False):
    if '</think>' in response:
        think_part  = response.split('</think>')[0].replace('<think>', '').strip()
        answer_part = response.split('</think>', 1)[-1].strip()
        if show_thinking:
            print('[思考過程]'); print('-'*60)
            for line in think_part.split('\n')[:20]:
                print(textwrap.fill(line, 80))
            if think_part.count('\n') > 20: print('...(省略)...')
            print()
        print('[回答]'); print('-'*60)
        print(answer_part.strip())
    else:
        print('[回答]'); print('-'*60)
        print(response.strip())


def mem(): print(f'RAM: {psutil.virtual_memory().used/1e9:.1f}GB')


# 接続テスト
t0 = time.time()
test = ask('こんにちは！一文で自己紹介してください。', think=False, max_tokens=60)
print(f'接続OK ({time.time()-t0:.0f}s): {test}')
mem()
```

## 5. 数学推論 (think mode)


```python
t0 = time.time()
resp = ask(
    '次の方程式を解いてください: 3x² - 12x + 9 = 0\n'  
    '解の公式を使って、全ての解を求めてください。',
    think=True, max_tokens=500,
)
print(f'Time: {time.time()-t0:.0f}s'); mem()
show(resp, show_thinking=True)
```

## 6. 論理パズル (think mode)


```python
t0 = time.time()
resp = ask(
    'A、B、C の3人がいます。\n'
    '・Aは正直者か嘘つきのどちらかです。\n'
    '・Bは「Aは嘘つきだ」と言いました。\n'
    '・Cは「Bは正直者だ」と言いました。\n'
    'A、B、Cはそれぞれ正直者ですか？論理的に説明してください。',
    think=True, max_tokens=600,
)
print(f'Time: {time.time()-t0:.0f}s'); mem()
show(resp, show_thinking=True)
```

## 7. コード生成 (nothink mode)


```python
t0 = time.time()
resp = ask(
    'Pythonで素数判定関数を書いてください。'  
    'エラトステネスの篩を使って1000以下の素数を全て列挙するコードも含めてください。',
    think=False, max_tokens=500,
)
print(f'Time: {time.time()-t0:.0f}s'); mem()
show(resp)
```

## 8. think vs nothink 比較


```python
question = '1から100までの整数の合計を求めてください。'

print('[think]')
t0 = time.time()
r_think = ask(question, think=True, max_tokens=300)
t_think = time.time() - t0

print('[nothink]')
t0 = time.time()
r_nothink = ask(question, think=False, max_tokens=100)
t_nothink = time.time() - t0

print('='*60)
print(f'think ({t_think:.0f}s)'); print('-'*60)
show(r_think, show_thinking=True)
print()
print('='*60)
print(f'nothink ({t_nothink:.0f}s)'); print('-'*60)
show(r_nothink)
print(f'\nthink は nothink の {t_think/t_nothink:.1f}倍 遅い')
```
